```python
# Line Plot
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1, color = 'red', label = 'Apple', marker = 'o') 
# if you give a single parameter then it will be considered as parameters of Y-axis and the indices will be X-axis
plt.plot(x,y2, color = 'blue', label = 'Kiwi', linestyle = 'dashed', marker ='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over time")
plt.legend()
plt.show() # plt.show  will draw every line before it on a single graph
```

# Scatter plot

```python
#adjust the size of figure of plot
plt.figure(figsize = (4,4))
plt.scatter(x,y1, color = 'red', label = 'Apple', marker = 'o') 
# if you give a single parameter then it will be considered as parameters of Y-axis and the indices will be X-axis
plt.scatter(x,y2, color = 'blue', label = 'Kiwi', linestyle = 'dashed', marker ='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over time")
plt.legend()
plt.show() # plt.show  will draw every line before it on a single graph
```

# Histogram

```python
xsn = np.random.randn(100)
sigma  = 10
u  = 80
X = np.round(xsn*sigma + u)
X2 = np.round(xsn*10 + 40)
print(X,X2)
```

```python
plt.hist(X, alpha = 0.8 , label = 'maths')
plt.hist(X2, alpha = 0.8, label = 'physics')
plt.xlabel('Marks Range')
plt.ylabel('prob/frequency of students getting that range')
plt.legend()
plt.show()
```

# Data visualization Challenge
Given a movies database you have to make a visualization that - 
- Plot the length of movie Title name of X-axis 
- Frequency counts on Y-axis ie Number of movies having 'x' characters in their title

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read csv file
df = pd.read_csv('movie_metadata.csv')
```

```python
df.head(10)
```

```python
titles = list(df.get('movie_title'))
```

```python
print(titles)
```

```python
freq_titles = {}

for title in titles:
    length = len(title)
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel('Length of movie titles')
plt.ylabel('frequency count')
plt.title('Movie data visualization Problem')
plt.show()
```
