# Creating Movie Recommendation

```python
df
```

```python
moviemat = df.pivot_table(index = 'user_id', columns='title', values = 'rating')
```

```python
moviemat
```

```python
ratings.sort_values('number of ratings', ascending=False).head()
```

```python
starwars_userRatings = moviemat['Star Wars (1977)']
starwars_userRatings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_userRatings)
```

```python
similar_to_starwars
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
```

```python
corr_starwars
```

```python
corr_starwars.dropna(inplace =True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation', ascending=False).head()
```

```python
# but perfect correlation is due to less number of users rated that movie, so we will set a threshold
ratings
```

```python
corr_starwars = corr_starwars.join(ratings['number of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['number of ratings']>100].sort_values('Correlation', ascending=False)
```

# Prediction Function

```python
def predict_movies(movie_name):
    movie_user_ratings = moviemat[movie_name]
    similar_to_movie = moviemat.corrwith(movie_user_ratings)
    
    corr_movie = pd.DataFrame(similar_to_movie, columns = ['Correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie = corr_movie.join(ratings['number of ratings'])
    predictions = corr_movie[corr_movie['number of ratings']>100].sort_values('Correlation', ascending=False)  
    
    return predictions
```

```python
predictions = predict_movies('Titanic (1997)')
```
