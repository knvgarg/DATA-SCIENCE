```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips') #Builtin function have datasets 
```

```python
tips.head()
```

```python
sns.barplot(x ='sex' ,y ='total_bill' ,data = tips, estimator=np.std)
```

# count plot 
if you want to count number of istances of a particular catagory
- there is no value y given as parameter

```python
sns.countplot(x = 'sex',data = tips)
```

```python
sns.countplot(x ='time', data = tips)
```

```python
# helps in visualizing mean, std_dev, min, max and OUTLIERS
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
# same as boxplot
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker', split = 'True')
```

# Distplot
to see how spread a given variable is

```python
plot = sns.displot(tips['total_bill'], bins=100, kde = 'True')
```

#### Kdeplot is a Kernel Distribution Estimation Plot which depicts the probability density function of the continuous or non-parametric data variables i.e. we can plot for the univariate or multiple variables altogether.

```python
# KDE
kde = sns.kdeplot(tips['total_bill'])
```

#### Seaborn's jointplot displays a relationship between 2 variables (bivariate) as well as 1D profiles (univariate) in the margins. This plot is a convenience class that wraps JointGrid.

```python
# JointPlot
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
# also check the kind attribute of this graph
```

# PairPlot

```python
sns.pairplot(tips, hue = 'sex')
```

#### A correlation is a statistical measurement of the relationship between two variables. Possible correlations range from +1 to 1. A zero correlation indicates that there is no relationship between the variables.

```python
tips_corr = tips.corr()
tips_corr
```

# Heatmap
- Takes parameter which have variables on both sides of the table (as the tips_corr)
- Light color means high corelation, dark color means low corelation

```python
sns.heatmap(tips_corr, annot =True)
```
